```python
import pandas as pd

# Load the CSV file (use correct path)
df = pd.read_csv("Twitter_Data.csv", encoding="utf-8", low_memory=False)

# âœ… Your dataset columns:
# 'clean_text' â†’ text content of tweet
# 'category'   â†’ sentiment label

TEXT_COL = 'clean_text'
LABEL_COL = 'category'

# Keep only required columns
df = df[[TEXT_COL, LABEL_COL]].dropna().reset_index(drop=True)

print("âœ… Data loaded successfully!")
print(df.head())
print("\nColumns:", df.columns.tolist())
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

STOPWORDS = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|@\w+|#","", text)
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['text_clean'] = df[TEXT_COL].apply(clean_text)
print("âœ… Text cleaned successfully!")
df[['clean_text','text_clean','category']].head()
```

```python
# --- Import the class first ---
from sklearn.preprocessing import LabelEncoder

# --- Encode labels (convert text sentiments to numeric values) ---
le = LabelEncoder()
df["label_encoded"] = le.fit_transform(df["category"])

# âœ… Confirm encoding worked
print("Encoded labels:", dict(zip(le.classes_, le.transform(le.classes_))))
```
